```python
''' 1. For-loops - Run de code en kijk wat er gebeurd als je 'appel' toevoegd'''
fruits = ['mango', 'banaan', 'druif']
user_input = input('Welk fruit wilt u?')
input_gevonden = False
for fruit in fruits:
  if fruit == user_input:
    input_gevonden = True
if input_gevonden:
  print('Er is een appel in de fruits-lijst')
else:
  print('Er is geen appel in de fruits-lijst')
```
