# Loops
## Introductie

Loops zijn heel handig als we bepaalde code meerdere keren willen uitvoeren.

We onderscheiden 2 soorten loops: **for** loops en **while** loops.

## For loops

```python
# For loops gebruiken we als we iets een aantal keer willen uitvoeren:
for x in range(5): # range 5 is het domain van 0 tot 5 dus 0, 1, 2, 3 en 4
    print(x)
```

```python
# For loops kunnen we ook gebruiken om iets uit te voeren voor elk item in een list:
for x in ["boer", "zoekt", "vrouw"]:
    print(x.upper())
```

```python
# Nog een voorbeeld:
ls1 = [5, 6, 7, 8]
for list_item in ls1:
    if list_item == 7:
        print("Er zit een 7 in deze lijst")
```

## While loops

```python
# While loops worden uitgevoerd net zolang het statement wat achter de while loop evalueert naar True:
i = 10
while i > 5:
    print("De waarde van i is nu:")
    print(i)
    i = i - 1
```

```python
# Dit is een voorbeeld van een while loop die nooit uitgevoerd wordt:
while False:
    print("Ik word nooit geprint!")
```

```python
# Of een oneindige loop, pas op met uitvoeren..!

from time import sleep # Dit is een import statement en komt later nog aan bod

while True:
    print("Dit is een oneindige loop!")
    sleep(1) # Dit zorgt ervoor dat het programma steeds 1 seconde wacht
    
# Tip: zit je vast in een oneindige loop, klik dan op de vierkante stopknop in de menubalk hierboven (in bijvoorbeeld Jupyter Lite)
```
